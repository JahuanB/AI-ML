```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score


X = np.array([1200, 1400, 1600, 1700, 1800, 1850, 1900, 2000, 2200, 2400]).reshape(-1, 1)
Y = np.array([220, 260, 290, 310, 330, 345, 360, 380, 410, 450])
```

```python
# @title Plot Model
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Data Points')
plt.xlabel('House Size (sq ft)')
plt.ylabel('Price (thousands $)')
plt.title('House Price vs. Size')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# @title Build Linear Regression Model
model = LinearRegression()
model.fit(X, Y)

y_pred = model.predict(X)
print(y_pred)
print(Y)
```

```python
# @title Plot Model Results
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, y_pred, color='red', linewidth=2, label='Linear Regression Line')
plt.xlabel('House Size (sq ft)')
plt.ylabel('Price (thousands $)')
plt.title('House Price vs. Size')
plt.grid(True)
plt.legend()
plt.show()

#print model parameters
print('Intercept:', model.intercept_)
print('Coefficient:', model.coef_)

```

```python
r2 = r2_score(Y, y_pred)
print(f'R-squared: {r2:4f}')
```

```python
new_size = np.array([[2100]])
predicted_price = model.predict(new_size)[0]
print(f'Predicted price for 2100 sq ft house: ${predicted_price:.2f}k')
```
